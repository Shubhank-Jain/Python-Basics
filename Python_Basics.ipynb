```python
# Make new bucket named unique_bucket and sort in ascending order using for loop, if & not, .append in function.
# Solution 2

bucket = [2,3,5,6,6,9,1,2]
unique_bucket = []
for i in bucket:
  if i not in unique_bucket:
    unique_bucket.append(i)
    unique_bucket.sort()
print(unique_bucket)
```
